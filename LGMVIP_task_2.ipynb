```python
x=np.array(df['High'])
y=np.array(df['Low'])
plt.bar(x,y)
plt.figure(figsize=(10,10))
plt.show()
```

```python
x=np.array(df['Turnover (Lacs)'])
y=np.array(df['Total Trade Quantity'])
plt.bar(x,y)
plt.xlabel('Turnover')
plt.ylabel('Quantity')
plt.show()
```

```python
df['Daily Return']=df['Close'].pct_change()
```

```python
plt.figure(figsize=(10,7))
df['Daily Return'].hist(bins=50)
plt.xlabel('Daily Return')
plt.title('COMPANY')
plt.show()
```

```python
df['Daily Return'].plot(color='g',marker='*')
```
